```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import plotly as px
import seaborn as sns


path="https://frenzy86.s3.eu-west-2.amazonaws.com/python/data/datalr.csv"
```

```python
points = np.genfromtxt(path, delimiter=',')

#Extract columns
x = np.array(points[:,0])
y = np.array(points[:,1])
```

```python
# x = 10 * generate_random.rand(100)

# noise = np.random.normal(0,1,100)
# y = 3 * x + noise
```

```python
X = x.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)
```

```python
plt.plot(X,y,"o")
plt.plot(X, y_pred, "-r")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2, 
                                                    random_state = 667)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test) #input che non ha mai visto
```

```python
length = y_pred.shape[0] #  
x = np.linspace(0,length,length)

plt.figure(figsize=(10,7))
plt.plot(x, y_test, label='real y')
plt.plot(x, y_pred, label="predicted y'")
plt.legend(loc=2);
```
